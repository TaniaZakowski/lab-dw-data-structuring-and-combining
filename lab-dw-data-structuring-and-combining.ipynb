```python
# Your code goes here
import pandas as pd
import numpy as nu

data_lab_Data22=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
data_lab_Data23=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
#data_lab_Data2_cleaned=pd.read_csv("..\lab-dw-data-cleaning-and-formatting\data_lab_Data2_cleaned.csv")
data_lab_Data2_cleaned=pd.read_csv("data_lab_Data2_cleaned.csv")
#display(data_lab_Data2.head(2))
display(data_lab_Data2_cleaned.head(2))
display(data_lab_Data22.head(2))
display(data_lab_Data23.head(2))
```

```python
dataframes = {
    'data_lab_Data2_cleaned': data_lab_Data2_cleaned,
    'data_lab_Data22': data_lab_Data22,
    'data_lab_Data23': data_lab_Data23
}
```

```python
#Change the column names for the 3 datasets have the same as the first
data_lab_Data22.columns = data_lab_Data22.columns.str.replace(' ', '_').str.lower()
data_lab_Data22.rename(columns={'st': 'state'}, inplace=True)
data_lab_Data23.columns = data_lab_Data23.columns.str.replace(' ', '_').str.lower()

#checking that all columns have the same name between datasets
for df_name, df in dataframes.items():
    print(f"DataFrame Name: {df_name}")
    #print(f"Shape: {df.shape}")
    print(f"Column Names: {df.columns.tolist()}")
    #print(f"Data Types:\n{df.dtypes}\n")
```

```python
#Cleaning invalid Values for consistency of columns from the 3 datasets
#Searching for values present in the datasets for relevant columns (for exemple, customers should have only unique values.
# List of columns to check for unique values
columns_to_check = ['state', 'gender', 'education', 'policy_type', 'vehicle_class']

# Loop through each DataFrame to print unique values of specified columns
for df_name, df in dataframes.items():
    print(f"Unique values in {df_name}:")
    for column in columns_to_check:
        if column in df.columns:
            unique_values = df[column].unique()
            print(f"{column}: {unique_values}")
```

```python
#Treating the gender column to keep 'Female" or "Male" as possible values
def categorize_gender(x):
    if isinstance(x, str):
        if x[0] in ['F', 'f']:
            return 'Female'
        elif x[0] in ['M', 'm']:
            return 'Male'
    return x

data_lab_Data23['gender'] = data_lab_Data23['gender'].apply(categorize_gender)
data_lab_Data22['gender'] = data_lab_Data22['gender'].apply(categorize_gender)

#checking:
for df_name, df in dataframes.items():
    print(f"Unique values in {df_name}:")
    for column in columns_to_check:
        if column in df.columns:
            unique_values = df[column].unique()
            print(f"{column}: {unique_values}")
```

```python
#Treating the state column 
def replace_state_names(df):
    replacements = {
        'WA': 'Washington',
        'Cali': 'California',
        'AZ': 'Arizona'
    }
    df['state'] = df['state'].replace(replacements)
    return df

data_lab_Data22 = replace_state_names(data_lab_Data22)
data_lab_Data23 = replace_state_names(data_lab_Data23)

#checking
for df_name, df in dataframes.items():
    print(f"Unique values in {df_name}:")
    for column in columns_to_check:
        if column in df.columns:
            unique_values = df[column].unique()
            print(f"{column}: {unique_values}")
```

```python
#Treating the education column 
def replace_bachelor(df):
    df['education'] = df['education'].replace('Bachelor', 'Bachelors')
    return df
data_lab_Data22 = replace_bachelor(data_lab_Data22)
data_lab_Data23 = replace_bachelor(data_lab_Data23)

#checking
for df_name, df in dataframes.items():
    print(f"Unique values in {df_name}:")
    for column in columns_to_check:
        if column in df.columns:
            unique_values = df[column].unique()
            print(f"{column}: {unique_values}")
```

```python
#Treating the vehicle_class column 
def replace_vehicle_class(df):
    replacements = {
        'Sports Car': 'Luxury',
        'Luxury SUV': 'Luxury',
        'Luxury Car': 'Luxury'
    }
    df['vehicle_class'] = df['vehicle_class'].replace(replacements)
    return df

data_lab_Data22 = replace_vehicle_class(data_lab_Data22)
data_lab_Data23 = replace_vehicle_class(data_lab_Data23)

#checking
for df_name, df in dataframes.items():
    print(f"Unique values in {df_name}:")
    for column in columns_to_check:
        if column in df.columns:
            unique_values = df[column].unique()
            print(f"{column}: {unique_values}")
```

```python
#Treating the columns customer_lifetime_value
#Checking for datatype 
for name, df in dataframes.items():
    if 'customer_lifetime_value' in df.columns:
        print(f"In DataFrame '{name}', the datatype of the column 'customer_lifetime_value' is: {df['customer_lifetime_value'].dtype}")
    else:
        print(f"In DataFrame '{name}', the column 'customer_lifetime_value' does not exist.")
```

```python
#remove the caracter % the columns customer_lifetime_value
def remove_percentage_sign(df):
    df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
    return df

data_lab_Data22 = remove_percentage_sign(data_lab_Data22)
#Change the dataype to fit the other datasets
data_lab_Data22['customer_lifetime_value'] = data_lab_Data22['customer_lifetime_value'].astype(float)
```

```python
#Changing the data in number_of_open_complaints for the dataset data_lab_Data22 , as data_lab_Data23 already has the right format
def transform_complaints(value):
    """
    Transform the value if it contains '/' and is a string.
    """
    if isinstance(value, str) and '/' in value:
        return value.split('/')[1]
    return value

data_lab_Data22['number_of_open_complaints'] = data_lab_Data22['number_of_open_complaints'].apply(transform_complaints)
data_lab_Data22['number_of_open_complaints'] = data_lab_Data22['number_of_open_complaints'].astype(float)
#data_lab_Data22.head()
data_lab_Data22.number_of_open_complaints.unique()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Create the pivot table
summary_table = data_mk_cust_analysis.pivot_table(
    index='sales_channel', 
    values='customer_lifetime_value', 
    aggfunc='sum' 
)

# Round the values to 2 decimal points
summary_table = summary_table.round(2)

# Print the summary table
print(summary_table)
```

```python
summary_table.describe()
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Create the pivot table
summary_table2 = data_mk_cust_analysis.pivot_table(
    index=['gender','education'],
    values='customer_lifetime_value', 
    aggfunc='mean' 
)

# Round the values to 2 decimal points
summary_table2 = summary_table2.round(2)

# Print the summary table
print(summary_table2)
```

```python
print(summary_table2.mean())
print(summary_table2.median())
```
